In this code, the initialize_grid function creates a random initial grid with a specified density of active cells. The evolve function updates the grid based on the stochastic rules defined by the probability kernel. If it is inactive (cell value of 0), it becomes active with a probability defined by the kernel.

The update function is called for each frame of the animation. It evolves the grid and updates the image object accordingly.

You can adjust the width, height, initial_density, and kernel variables to customize the model according to your requirements. The frames parameter in animation.FuncAnimation controls the number of frames in the animation, and the interval parameter controls the delay between frames in milliseconds.

Feel free to modify the code to experiment with different probability kernels or visualization settings.

```python
# Configuration
width = 100
height = 100
initial_density = 0.3
kernel = np.ones((3, 3)) * 0.25  # Example kernel where each neighbor has equal probability

# Initialize grid and animation
grid = initialize_grid(width, height, initial_density)
fig, ax = plt.subplots()
img = ax.imshow(grid, cmap='Greys', interpolation='none')
ani = animation.FuncAnimation(fig, update, frames=200, interval=100, blit=True)

plt.show()
```
